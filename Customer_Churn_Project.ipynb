```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"
    
df["SeniorCitizen"]= df["SeniorCitizen"].apply(conv)
```

```python
# Calculate the percentages
# This groups by 'SeniorCitizen' and calculates normalized (percentage) value counts for 'Churn'
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack()

# Plotting the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 4))  # Adjust the figure size for better visualization
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])  # Customize colors

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height * 100:.1f}%', 
            ha='center', va='center')  # Show percentage with 1 decimal point

# Customize the chart
plt.title("Churn by Senior Citizen (Stacked Bar Chart)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)  # Adjust rotation if needed
plt.legend(title="Churn", loc="upper right")  # Customize legend location

plt.show()  # Display the plot
```

```python
# List of the columns you want to create count plots for
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create a subplot with a grid (adjust the number of rows and columns as needed)
n_cols = 3  # Number of columns in the grid
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate number of rows

fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 12))
axes = axes.flatten()  # Flatten the axes array to make indexing easier

# Loop over each column and plot the countplot in the corresponding axis
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i],hue="Churn")
    axes[i].set_title(f'Count Plot: {col}')
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45, ha='right')  # Rotate x-axis labels for readability

# Remove any empty subplots if there are more axes than plots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()  # Adjust the layout for better spacing
plt.show()
```
